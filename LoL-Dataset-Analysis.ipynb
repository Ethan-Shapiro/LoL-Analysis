<h1>EDA</h1>

<p>Firstly, let's understand what each attribute means</p>
<ol>
    <li><b>Wards</b> - Wards can be placed by players to reveal areas of the map that were previously dark. They can be destryoed using various items in the game </li> 
    <li><b>First Blood</b> - First bloods contribute 400 gold to player(s), giving the player a lead from the start</li>
    <li><b>KDA</b> - How many times a player was killed/dead or players asissted in killing the enemy.</li>
    <li><b>Dragons</b> - Dragons give powerful buffs to the team which slays them and therefore have a lot of influence over the outcome of the game.</li>
    <li><b>Towers Destroyed</b> - Destroying a tower opens up the map to allow the team to move into more areas.</li>
    <li><b>Gold</b> - Having more gold than your opponent allows you to buy more items and therefore be more powerful.</li>
    <li><b>Experience</b> - The more experience someone has, the higher the level and more powerful they are.</li>
</ol>
<p>Now that we have a general understanding of each feature and its affect on the outcome of the game, let's look into each teams metrics<p>

<h1>Model Selection</h1>

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

X = df_clean
y = target

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

<h3>Naive Bayes</h3>

```python
from sklearn.naive_bayes import GaussianNB

bayes = GaussianNB()

bayes.fit(X_train, y_train)
pred = bayes.predict(X_test)

print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

<h3>Decision Tree</h3>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dt = DecisionTreeClassifier()

# search the best params
grid = {'min_samples_split': [5, 10, 20, 50, 100], 'max_leaf_nodes': [10, 50, 100], 'criterion': ['gini', 'entropy']}

grid_tree = GridSearchCV(dt, grid, cv=5)
grid_tree.fit(X_train, y_train)

pred = grid_tree.predict(X_test)

print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

<h3>Random Forest</h3>

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

grid = {'criterion': ['gini', 'entropy'], 'max_depth': [2, 5, 10], 'n_estimators': [100, 250, 500]}

grid_rf = GridSearchCV(rf, grid, cv=5)
grid_rf.fit(X_train, y_train)

pred = grid_rf.predict(X_test)

print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

<h3>Logistic Regression</h3>

```python
from sklearn.linear_model import LogisticRegression

lm = LogisticRegression()

lm.fit(X_train, y_train)

pred = lm.predict(X_test)

print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

<h3>K-nearest Neighbors</h3>

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

grid = {'n_neighbors': np.arange(1, 50)}
grid_knn = GridSearchCV(knn, grid, cv=5)
grid_knn.fit(X_train, y_train)

pred = grid_knn.predict(X_test)

print(accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

Logistic regression achieved the best scores. Let's do some PCA to try and improve the scores
