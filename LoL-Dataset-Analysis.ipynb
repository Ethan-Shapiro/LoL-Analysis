<h1>EDA</h1>

<p>Firstly, let's understand what each attribute means</p>
<ol>
    <li><b>Wards</b> - Wards can be placed by players to reveal areas of the map that were previously dark. They can be destryoed using various items in the game </li> 
    <li><b>First Blood</b> - First bloods contribute 400 gold to player(s), giving the player a lead from the start</li>
    <li><b>KDA</b> - How many times a player was killed/dead or players asissted in killing the enemy.</li>
    <li><b>Dragons</b> - Dragons give powerful buffs to the team which slays them and therefore have a lot of influence over the outcome of the game.</li>
    <li><b>Towers Destroyed</b> - Destroying a tower opens up the map to allow the team to move into more areas.</li>
    <li><b>Gold</b> - Having more gold than your opponent allows you to buy more items and therefore be more powerful.</li>
    <li><b>Experience</b> - The more experience someone has, the higher the level and more powerful they are.</li>
</ol>
<p>Now that we have a general understanding of each feature and its affect on the outcome of the game, let's look into each teams metrics<p>

```python
# We have an even data set woooh!
sns.countplot(x=df['blueWins'])
plt.show()
print(df['blueWins'].value_counts())
```

```python
#Drop repeated columns (Red Deaths = Blue Kills, )
cols = ['redHeralds', 'redDragons', 'redGoldDiff', 'redExperienceDiff', 'blueCSPerMin', 'redCSPerMin', 'blueGoldPerMin',
        'redGoldPerMin', 'redTotalMinionsKilled', 'blueTotalMinionsKilled', 'redKills', 'redDeaths', 
        'redFirstBlood', 'gameId']

df_clean = df.drop(cols, axis=1)
df_clean.columns
```

```python
plt.figure(figsize=(18, 14))
sns.heatmap(df_clean.corr(), cmap='YlGnBu', annot=True, fmt='.2f', vmin=0)
```

```python
# Remove features with high colinearity
cols = ['blueTotalGold', 'blueAssists', 'redTotalGold', 'redTotalExperience', 'redAvgLevel', 'redAssists', 
        'blueAvgLevel', 'blueAvgLevel']
df_clean = df_clean.drop(cols, axis=1)
```

```python
plt.figure(figsize=(16, 12))
sns.heatmap(df_clean.corr(), cmap='YlGnBu', annot=True, fmt='.2f', vmin=0)
```

```python
# Remove features that don't have a correlation with the target, blueWins
corr = df_clean.corr()['blueWins'][1:]
target = df_clean['blueWins']
i = 0
cols = []
for c in corr:
    if c < -.2 or c > .2:
        cols.append(df_clean.columns[i])
    i += 1

df_clean = df_clean.loc[:, cols]
```

```python
df_clean.columns
```

```python
plt.figure(figsize=(16, 12))
sns.heatmap(df_clean.corr(), cmap='YlGnBu', annot=True, fmt='.2f', vmin=0)
```

```python
df_clean.hist(alpha=.7, bins=5, figsize=(12, 10))
plt.show()
```

```python
ax = sns.PairGrid(pd.concat([df_clean, target], axis=1), 
                  vars=['blueWardsDestroyed', 'blueFirstBlood', 'blueKills', 'blueDeaths'], 
                  hue='blueWins', palette='coolwarm')
ax.map_diag(plt.hist)
ax.map_offdiag(plt.scatter)
plt.show()
```
